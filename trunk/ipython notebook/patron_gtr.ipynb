```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from scipy.misc import imread
import datetime as dt
import math
```

```python
def pattern(lats,lons):
    N = len(lats)-1
    pt = []
    param = 0
    lat = 0
    lon = 0
    for i in range(N):
        lat = lats[i]
        lon = lons[i]
        next_lat = lats[i + 1]
        next_lon = lons[i + 1]
        param = math.atan2(next_lat - lat,next_lon - lon)
        #(lat*next_lat + lon*next_lon) / math.sqrt((lat*lat + lon*lon)*(next_lat*next_lat + next_lon*next_lon))
        pt.append(param)
    return pt
```

```python
file_csv = '/home/flopez/SVN/gitrd/trunk/ipython notebook/gtr_lite_christian.csv'

data= pd.read_csv(file_csv, delimiter=';', quotechar='"', 
                  usecols=[2,3,4], parse_dates=[2], 
                  names=['LAT','LON','TIME'], dtype = {2:np.float64,3:np.float64})
print data.dtypes
```

```python
d = []
for i in range(7):
    d.append(data.ix[(data['TIME'] > dt.datetime(2014,2,i+2,0,0,0)) & ((data['TIME'] < dt.datetime(2014,2,i+2,23,59,59)))])
```

```python
lat, lon = [], []
for dias in d:
    lat.append(dias.LAT.values)
    lon.append(dias.LON.values)
```

```python
pt = []
for i in range(7):
    pt.append(pd.Series(pattern(lat[i],lon[i]),index=d[i]['TIME'][:-1]))
    
```
