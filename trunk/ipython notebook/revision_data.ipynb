```python
data_no_filter = pd.read_csv(url_no_filter, converters={1:f})
data_netmaster = pd.read_csv(url_netmaster, sep=';')
```

```python
def parserDate(date_string, year, mes, dia, hora):
    date_list = date_string.split()
    return datetime.datetime(int(date_list[year]), int(months.index(date_list[mes])), int(date_list[dia]),int(date_list[hora][0:2]),int(date_list[hora][3:5]),int(date_list[hora][6:8]))
```

```python
for index, date in data_no_filter['Event Time'].iteritems():
    data_no_filter['Event Time'][index] = parserDate(date,5,1,2,3)
```

```python
data_no_filter = data_no_filter.sort(columns=['Event Time'], ascending=True)
probable_cause_no_filter = Series(Series(data_no_filter['Probable Cause'].ravel()).unique())
dictionary = {}
for a in probable_cause_no_filter:
       dictionary[a] = {'Warning':[], 'Minor':[],'Major':[] , 'Critical':[]}
```

```python
for row in data_no_filter.iterrows():
    dictionary[row[1]['Probable Cause']][row[1]['Severity']].append(row[1]['Event Time'])
severity_probable_cause = pd.DataFrame(dictionary)
```

# PLOTING BY WARNING

```python
def severityNumber(number_instaces_vector):
    list_instances = []
    number=4
    for i in number_instaces_vector:
        list_instances.extend([number]*i)
        number = number - 1
    return list_instances
```

```python
def returnAxes(falla):
    vector_repeticiones = [len(b) for b in severity_probable_cause[falla]]
    vector_severity = severityNumber(vector_repeticiones)
    vector_dates = [x for a in severity_probable_cause[falla] for x in a]
    vector_to_plot = []
    vector_to_plot.append(vector_severity)
    vector_to_plot.append(vector_dates)
    return vector_to_plot
```
